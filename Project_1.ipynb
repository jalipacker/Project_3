```python
covid = []
with open('COVID-19_Daily_Counts_of_Cases__Hospitalizations__and_Deaths_20241107.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        covid.append(row)

covid[:3]
```

```python
cases = []
for row in covid: 
    cases.append(float(row['CASE_COUNT']))
```

#### Calculate the mean, median, and mode. Code sourced from: https://www.geeksforgeeks.org/finding-mean-median-mode-in-python-without-libraries/

```python
# calculate the mean
total_cases = sum(cases)
count = len(cases)
mean_cases = total_cases / count
print("Mean:", mean_cases)
```

```python
# calculate the median - first sort the list of numbers
n = len(cases) 
cases.sort() 
cases
```

```python
# calculate the median - check if the middle number is even or odd and proceed accordingly 
if n % 2 == 0: 
    median1 = cases[n//2] 
    median2 = cases[n//2 - 1] 
    median = (median1 + median2)/2
else: 
    median = cases[n//2] 
print("Median is: " + str(median)) 
```

```python
# calculate the mode, code sourced from chatGPT: https://chatgpt.com/share/67393147-9d7c-8008-aec9-6e705d8f48e5

frequency_dict = {}
for value in cases:
    if value in frequency_dict:
        frequency_dict[value] += 1
    else:
        frequency_dict[value] = 1
```

```python
# find values with max frequency 
max_frequency = max(frequency_dict.values())
```

```python
# use multimode as we know from Step 3 that there are two modes for this data
modes = [key for key, count in frequency_dict.items() if count == max_frequency]
```

```python
print(f"The mode(s) is/are: {modes}")
```

## Data visualization

```python
#read in the covid dataset again using pandas and view first five rows
covid = pd.read_csv('COVID-19_Daily_Counts_of_Cases__Hospitalizations__and_Deaths_20241107.csv')
covid.head()
```

```python
# convert date to datetime and extract year from the date - code for datetime sourced from chatGPT
covid['date_of_interest'] = pd.to_datetime(covid['date_of_interest'])
covid['year'] = covid['date_of_interest'].dt.year
```

```python
# calculate the average case count by year
average_case_count_by_year = covid.groupby('year')['CASE_COUNT'].mean()
```

```python
# Generate a data visualization for the average case count by year
data_viz = "Average Daily COVID-19 Case Count by Year\n" 
for year, avg_count in average_case_count_by_year.items():
    stars = '*' * round(avg_count / 100)  # Each '*' represents approximately 100 cases
    data_viz += f"{year}: {stars} ({avg_count:.0f} cases)\n"

data_viz += "=" * 50 + "\nNote: Each '*' represents approximately 100 cases."
print(data_viz)
```
